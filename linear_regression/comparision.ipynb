In this notebook :
- Generate synthetic data.
- Train a custom Linear Regression model implemented from scratch.
- Visualize the fitted regression line.
- Plot the cost function over iterations.
- Compare results with scikit-learn.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from implementation import SimpleLinearRegression
```

```python
np.random.seed(42)

X=2*np.random.rand(100)
y=4+3*X+np.random.randn(100)
```

Visualize the Dataset

```python
plt.figure(figsize=(6,4))
plt.scatter(X,y,alpha=0.7)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Generated Data")
plt.show()
```

```python
model=SimpleLinearRegression(learning_rate=0.01,n_iterations=1000)
```

```python
model.fit(X,y)
```

```python
print("Cofficient(m):",model.weight)
print("Intercept(b): ",model.bias)
```

```python
y_pred=model.predict(X)
```

Plot Regression Line

```python
plt.figure(figsize=(6,4))
plt.scatter(X,y,alpha=0.7,label="Actual Data")
plt.plot(X,y_pred,color="blue",label="Predicted Line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()
```

Compare with Scikit_Learn

```python
sk_model=LinearRegression()
sk_model.fit(X.reshape(-1,1),y)
sk_pred=sk_model.predict(X.reshape(-1,1))


print("Sklearn Coefficient(m): ",sk_model.coef_)
print("Sklearn Intercept(b): ",sk_model.intercept_)
```

```python

plt.figure(figsize=(6, 4))
plt.scatter(X, y, alpha=0.7, label="Data")
plt.plot(X, y_pred, label="Custom Model")
plt.plot(X, sk_pred, linestyle="--", label="Scikit-Learn")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Custom vs Scikit-Learn")
plt.legend()
plt.show()
```
